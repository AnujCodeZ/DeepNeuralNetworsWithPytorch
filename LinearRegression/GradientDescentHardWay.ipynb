```python
import torch
```

```python
w = torch.tensor(10.0, requires_grad=True) # We want w to be modified. So grad is true.
X = torch.arange(-3,3,0.1).view(-1,1) # Features.
f = -3*X # Function.
```

```python
import matplotlib.pyplot as plt
plt.plot(X.numpy(),f.numpy())
plt.show()
```

```python
Y = f+0.1*torch.randn(X.size()) # Adding some noise.
plt.plot(X.numpy(),Y.numpy(),'ro')
plt.show()
```

```python
def forward(x): # Prediction function.
    return w*x

def criterion(yhat, y): # This is the cost function.
    return torch.mean((yhat-y)**2)
```

```python
lr = 0.1 # Learning rate.
COST = []
for epoch in range(4):
    yhat = forward(X)
    loss = criterion(yhat, Y)
    loss.backward() # Calculates gradients w.r.t. variables whose requires_grad = True.
    w.data = w.data - lr*w.grad.data
    
    w.grad.data.zero_() # Required because it will modify iteratively.
    
    COST.append(loss.item()) # To have the record of cost.
```

```python
niter = [1,2,3,4] # Number of iterations.
plt.plot(niter, COST)
plt.show()
```
